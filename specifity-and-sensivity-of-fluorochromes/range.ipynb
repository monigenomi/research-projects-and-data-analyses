```python
specificity_data = [sybrsafe_specificity,
                        dnaq_specificity,
                    picogreen_specificity,
                    evagreen_specificity,
            ethydium_bromide_specificity]
specificity_data_names = ["sybrsafe_specificity",
                              "dnaq_specificity",
                          "picogreen_specificity",
                            "evagreen_specificity",
                    "ethydium_bromide_specificity"]

def take_specificity(specificity_data, x):
    data = []
    for name in np.array(specificity_data):
        data.append(name[x])
    return data

rna_specificity = take_specificity(specificity_data, 0)
ssdna_specificity = take_specificity(specificity_data, 1)
protein_specificity = take_specificity(specificity_data, 2)

comparison_table = {
    "specificity_data_names": specificity_data_names,
    "specificity with RNA presence": rna_specificity,
    "specificity with ssDNA presence": ssdna_specificity,
    "Specificity with potein presence ": protein_specificity
    
}
pd.DataFrame(comparison_table)
```
